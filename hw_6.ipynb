```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
data_2d = pd.read_csv('/content/data_2d.csv')
mnist = pd.read_csv('/content/mnist.csv')
```

Завдання 1

Для data_2d.csv

```python
X_2d = data_2d.values
X_mnist = mnist.values[:, 1:]

scaler_2d = StandardScaler()
scaler_mnist = StandardScaler()
X_2d_scaled = scaler_2d.fit_transform(X_2d)
X_mnist_scaled = scaler_mnist.fit_transform(X_mnist)

k_values_2d = range(2, 11)
inertias_2d = []

for k in k_values_2d:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_2d_scaled)
    inertias_2d.append(kmeans.inertia_)
```

```python
plt.plot(k_values_2d, inertias_2d, 'bo-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Інерція')
plt.title('Метод ліктя для визначення оптимальної кількості кластерів (data_2d.csv)')
plt.show()
```

```python
optimal_k_2d = 4
kmeans_2d = KMeans(n_clusters=optimal_k_2d, random_state=42)
kmeans_2d.fit(X_2d_scaled)
labels_2d = kmeans_2d.labels_
```

```python
print("Результати кластеризації (data_2d.csv):")
print(labels_2d)
```

для mnist.csv

```python
k_values_mnist = range(2, 11)
inertias_mnist = []

for k in k_values_mnist:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_mnist_scaled)
    inertias_mnist.append(kmeans.inertia_)
```

```python
plt.plot(k_values_mnist, inertias_mnist, 'bo-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Інерція')
plt.title('Метод ліктя для визначення оптимальної кількості кластерів (mnist.csv)')
plt.show()
```

```python
optimal_k_mnist = 5  
kmeans_mnist = KMeans(n_clusters=optimal_k_mnist, random_state=42)
kmeans_mnist.fit(X_mnist_scaled)
labels_mnist = kmeans_mnist.labels_
```

```python
print("Результати кластеризації (mnist.csv):")
print(labels_mnist)
```
